```python
df = pd.read_csv("../input/playground-series-s3e24/train.csv")
df.head(5)
df.tail(5)
```

```python
# 正解データ
# 値
df["smoking"].unique()
# 出現回数
df["smoking"].value_counts()
```

```python
# 欠損値
df.isnull()
# 列単位で欠損値が存在するか調べる
df.isnull().any(axis = 0) # 行方向で調べるときはaxis = 1
# 各列の欠損値の個数
df.isnull().sum()
# 欠損値なかったのでp147~153省略
```

```python
# 特徴量と正解データを変数に代入
df.columns
x = df.loc[:, "age":"dental caries"]
# y = df["smoking"] どっちでも良さそう？
y = df.loc[:, "smoking"]
```

```python
# 決定木分析
from sklearn import tree
# モデルの作成
model = tree.DecisionTreeClassifier(max_depth = 45, random_state = 0)
# モデルの学習
model.fit(x, y)
# 学習済みモデルの正解率計算
model.score(x, y)
```

```python
# テストデータ
testdf = pd.read_csv("../input/playground-series-s3e24/test.csv")

test = testdf.loc[:, "age":"dental caries"]
test.head(5)
```

```python
res = model.predict(test)
# エラーで詰まったからここまで

output = pd.DataFrame({'id': test['id'], 'smoking': res})

# CSVファイルとして出力
output.to_csv('submission.csv', index=False)
```
